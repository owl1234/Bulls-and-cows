```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df=pd.read_csv('train.csv',usecols=['comment','reting'])
df.head()
```

```python
[x>4.5 for x in df.reting.values]
```

```python
comments=df.comment
sentiment=np.array([x>4.5 for x in df.reting]) # преобразовать в numpy.array из True и false, в котором true стоит,
           # если reting для комментария больше 4.5
assert type(sentiment)==np.ndarray
assert sentiment.dtype=='bool'
train,test,y_train,y_test=train_test_split(comments,sentiment,random_state=42,stratify=sentiment)
```

```python
common_words=train.str.lower().str.split(expand=True).stack().value_counts()
common_words
```

```python
l=common_words.index[:10000].tolist()
l
```

```python
l = common_words.index[:10000].tolist()
set_ = {word:index for index,word in enumerate(l)}
def bag_of_words(comment,list_of_words):
  """
  Сконвертировать строку в bag-of-words представление:
  подсчитать и вернуть в виде списка число раз,
  которое встречается в comment каждое слово из list_of_words
  """
  line = comment.lower().split()  
  res = np.zeros(len(list_of_words))
  
  for i in set(line):
    if i in list_of_words:
      res[set_[i]] += line.count(i)
  return res
```

```python
train_bows=train.apply(lambda c:bag_of_words(c,l)).tolist()
test_bows=test.apply(lambda c:bag_of_words(c,l)).tolist()
```

```python
X_train=np.array(train_bows)
X_test=np.array(test_bows)
assert X_train.shape[1]==X_test.shape[1]
assert X_train.shape[0]==train.shape[0]
assert X_test.shape[0]==test.shape[0]
```

```python
logreg=LogisticRegression(random_state=42)
logreg.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
def classification_results(y_test,y_pred):
  """ 
  По верным меткам и предсказаниям подсчитать значения
  True Positive, False Positive, False Negative, True Negative
  (вернуть кортеж из 4 элементов)
  """
  tp, fp, fn, tn = 0, 0, 0, 0
  for i in range(len(y_test)):
    if y_test[i] == 0 and y_pred[i] == 0:
      tn+=1
    if y_test[i] == 0 and y_pred[i] == 1:
      fp+=1
    if y_test[i] == 1 and y_pred[i] == 1:
      tp+=1
    if y_test[i] == 1 and y_pred[i] == 0:
      fn+=1
  
  return tp,fp, fn, tn
```

```python
assert classification_results([0,0,1,1,0,1,0],[0,1,1,1,1,0,1]) == (2,3,1,1)
assert classification_results([1,0,1,0],[1,0,1,0]) == (2,0,0,2)
```

```python
def accuracy(tp,fp,fn,tn):
  return (tp + tn) / (tp + tn + fn + fp)

def precision(tp,fp,fn,tn):
  return tp / (tp + fp)

def recall(tp,fp,fn,tn):
  return tp / (tp + fn)

assert accuracy(*classification_results([0,0,1,1,0,1,0],[0,1,1,1,1,0,1])) == accuracy_score([0,0,1,1,0,1,0],[0,1,1,1,1,0,1])
assert precision(*classification_results([0,0,1,1,0,1,0],[0,1,1,1,1,0,1])) == precision_score([0,0,1,1,0,1,0],[0,1,1,1,1,0,1])
assert recall(*classification_results([0,0,1,1,0,1,0],[0,1,1,1,1,0,1])) == recall_score([0,0,1,1,0,1,0],[0,1,1,1,1,0,1])
```

```python
test_predictions=logreg.predict(X_test)
clf_results=classification_results(y_test,test_predictions)
print('Accuracy: ',accuracy(*clf_results))
print('Precision: ',precision(*clf_results))
print('Recall: ',recall(*clf_results))
# можно ли улучшить качество?
# <your code possibly here>
```
